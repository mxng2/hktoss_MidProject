```python
FILE_PATH1 = "/content/drive/MyDrive/한국경제_TossBank/중간_프로젝트/data/ui_df.csv"
FILE_PATH2 = "/content/drive/MyDrive/한국경제_TossBank/중간_프로젝트/data/t_df.csv"
FILE_PATH3 = "/content/drive/MyDrive/한국경제_TossBank/중간_프로젝트/data/t_df.csv"
```

```python
import pandas as pd

df1 = pd.read_csv(FILE_PATH1)
df1.head()
```

```python
df2 = pd.read_csv(FILE_PATH2)
df2.head()
```

```python
df3 = pd.read_csv(FILE_PATH3)
df3.head()
```

```python
# '발급회원번호'를 기준으로 데이터 병합
merged_df = df1.merge(df2, on='발급회원번호', how='left').merge(df3, on='발급회원번호', how='left')
```

```python
# 병합된 데이터 확인
display(merged_df.head())
```

```python
# 결측치 확인
missing_values = merged_df.isnull().sum()
print("결측치가 있는 컬럼:\n", missing_values[missing_values > 0])
```

```python
# 데이터 타입 확인
print(merged_df.dtypes)
```

```python
# 중복된 행 확인
duplicates = merged_df.duplicated()
print(f"중복된 행의 수: {duplicates.sum()}")

# 열 이름이 '_x', '_y'로 끝나는 경우 중복된 열일 가능성 있음
duplicated_columns = [col for col in merged_df.columns if col.endswith('_x') or col.endswith('_y')]
print(f"중복 가능성이 있는 열: {duplicated_columns}")
```

```python
# 범주형 변수 확인
categorical_columns = merged_df.select_dtypes(include=['object']).columns
print("범주형 변수:\n", categorical_columns)
```

```python
# 병합된 데이터프레임을 CSV로 저장
output_file_path = "/path/to/your/directory/merged_data.csv"
merged_df.to_csv(output_file_path, index=False)

print(f"병합된 데이터가 {output_file_path}에 저장되었습니다.")
```
