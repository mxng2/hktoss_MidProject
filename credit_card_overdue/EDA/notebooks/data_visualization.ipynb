## 연령에 따른 연체자 분포 및 비율



```python
df_filtered = df[(df['잔액_B0M'] == 1) | (df['잔액_B0M'] == 2)]

age_columns = ['연령_20', '연령_30', '연령_40', '연령_50', '연령_60', '연령_70']

age_distribution = df_filtered[age_columns].sum().reset_index()
age_distribution.columns = ['연령대', '사람 수']

age_distribution['연령대'] = age_distribution['연령대'].apply(lambda x: x.split('_')[1] + '대')

fig = px.bar(age_distribution, x='연령대', y='사람 수',
             title='연령에 따른 연체자 분포',
             labels={'사람 수': '사람 수', '연령대': '연령대'},
             color='연령대')

fig.update_layout(xaxis_tickangle=-45)
fig.show()
```

```python
balance_columns = ['잔액_B0M']

age_groups = ['연령_20', '연령_30', '연령_40', '연령_50', '연령_60', '연령_70']
result = []

for age_group in age_groups:
    df_age_group = df[df[age_group] == 1]

    balance_distribution = df_age_group['잔액_B0M'].value_counts(normalize=True) * 100
    balance_distribution = balance_distribution.reset_index()
    balance_distribution.columns = ['잔액_B0M', '비율']
    balance_distribution['연령대'] = age_group

    result.append(balance_distribution)

balance_distribution_df = pd.concat(result)

fig = px.bar(balance_distribution_df,
             x='연령대',
             y='비율',
             color='잔액_B0M',
             title='연령대별 잔액_B0M 비율',
             labels={'비율': '비율 (%)', '연령대': '연령대', '잔액_B0M': '잔액_B0M'},
             barmode='stack')

fig.update_layout(xaxis_tickangle=-45)
fig.show()
```

## 20대 연체 여부에 따른 1순위 업종 비율 비교

```python
df_20s = df[df['연령_20'] == 1]

# '_1순위업종' 관련 컬럼들만 선택
first_priority_columns = [
    '_1순위업종_교육', '_1순위업종_교통', '_1순위업종_납부', '_1순위업종_사교활동',
    '_1순위업종_쇼핑', '_1순위업종_여유생활', '_1순위업종_요식', '_1순위업종_의료',
    '_1순위업종_일상생활', '_1순위업종_해외'
]

# 1순위 업종 비율 계산 함수
def calculate_priority_ratio(df):
    priority_distribution = df[first_priority_columns].sum().reset_index()
    priority_distribution.columns = ['업종', '빈도수']
    total = priority_distribution['빈도수'].sum()
    priority_distribution['비율'] = (priority_distribution['빈도수'] / total) * 100
    priority_distribution['업종'] = priority_distribution['업종'].apply(lambda x: x.replace('_1순위업종_', ''))
    return priority_distribution

# 잔액_B0M이 0, 1, 2인 사람들을 각각 필터링
df_20s_balance_0 = df_20s[df_20s['잔액_B0M'] == 0]
df_20s_balance_1 = df_20s[df_20s['잔액_B0M'] == 1]
df_20s_balance_2 = df_20s[df_20s['잔액_B0M'] == 2]

# 비율 계산
balance_0_distribution = calculate_priority_ratio(df_20s_balance_0)
balance_0_distribution['그룹'] = '미연체자'

balance_1_distribution = calculate_priority_ratio(df_20s_balance_1)
balance_1_distribution['그룹'] = '100만원 이하 연체자'

balance_2_distribution = calculate_priority_ratio(df_20s_balance_2)
balance_2_distribution['그룹'] = '100만원 초과 연체자'

# 세 그룹의 데이터를 결합
combined_distribution = pd.concat([balance_0_distribution, balance_1_distribution, balance_2_distribution])

# Plotly로 비율 시각화
fig = px.bar(combined_distribution,
             x='업종',
             y='비율',
             color='그룹',
             barmode='group',
             title='20대 연체 그룹에 따른 1순위 업종 비율 비교',
             labels={'비율': '비율 (%)', '업종': '1순위 업종'})

# 레이아웃 설정
fig.update_layout(xaxis_tickangle=-45)
fig.show()
```

## 20대 연체 그룹에 따른 1순위 쇼핑업종 비율

```python
df_20s = df[df['연령_20'] == 1]

df_balance_0 = df_20s[df_20s['잔액_B0M'] == 0]
df_balance_1 = df_20s[df_20s['잔액_B0M'] == 1]
df_balance_2 = df_20s[df_20s['잔액_B0M'] == 2]

first_priority_shopping_columns = [
    '_1순위쇼핑업종_도소매', '_1순위쇼핑업종_마트', '_1순위쇼핑업종_백화점',
    '_1순위쇼핑업종_쇼핑기타', '_1순위쇼핑업종_슈퍼마켓', '_1순위쇼핑업종_아울렛',
    '_1순위쇼핑업종_없음', '_1순위쇼핑업종_온라인', '_1순위쇼핑업종_편의점'
]

def calculate_shopping_ratio(df):
    distribution = df[first_priority_shopping_columns].sum().reset_index()
    distribution.columns = ['쇼핑업종', '빈도수']
    total = distribution['빈도수'].sum()
    distribution['비율'] = (distribution['빈도수'] / total) * 100
    return distribution

balance_0_distribution = calculate_shopping_ratio(df_balance_0)
balance_0_distribution['그룹'] = '미연체자'

balance_1_distribution = calculate_shopping_ratio(df_balance_1)
balance_1_distribution['그룹'] = '100만원 이하 연체자'

balance_2_distribution = calculate_shopping_ratio(df_balance_2)
balance_2_distribution['그룹'] = '100만원 초과 연체자'

combined_distribution = pd.concat([balance_0_distribution, balance_1_distribution, balance_2_distribution])

combined_distribution['쇼핑업종'] = combined_distribution['쇼핑업종'].apply(lambda x: x.replace('_1순위쇼핑업종_', ''))

fig = px.bar(combined_distribution,
             x='쇼핑업종',
             y='비율',
             color='그룹',
             barmode='group',
             title='20대 연체 그룹에 따른 1순위 쇼핑업종 비율',
             labels={'비율': '비율 (%)', '쇼핑업종': '쇼핑업종'})

fig.update_layout(xaxis_tickangle=-45)
fig.show()
```

## 20대 연체자와 나머지 연령 그룹 연체자 1순위 쇼핑 업종 비율 비교

```python
first_priority_shopping_columns = [
    '_1순위쇼핑업종_도소매', '_1순위쇼핑업종_마트', '_1순위쇼핑업종_백화점',
    '_1순위쇼핑업종_쇼핑기타', '_1순위쇼핑업종_슈퍼마켓', '_1순위쇼핑업종_아울렛',
    '_1순위쇼핑업종_없음', '_1순위쇼핑업종_온라인', '_1순위쇼핑업종_편의점'
]

df_balance_1_or_2 = df[df['잔액_B0M'].isin([1, 2])]

df_20s_balance_1_or_2 = df[(df['연령_20'] == 1) & (df['잔액_B0M'].isin([1, 2]))]

distribution_all = df_balance_1_or_2[first_priority_shopping_columns].sum().reset_index()
distribution_20s = df_20s_balance_1_or_2[first_priority_shopping_columns].sum().reset_index()

distribution_all.columns = ['쇼핑업종', '빈도수']
distribution_20s.columns = ['쇼핑업종', '빈도수']

total_count_all = distribution_all['빈도수'].sum()
total_count_20s = distribution_20s['빈도수'].sum()

distribution_all['비율'] = (distribution_all['빈도수'] / total_count_all) * 100
distribution_20s['비율'] = (distribution_20s['빈도수'] / total_count_20s) * 100

distribution_all['그룹'] = '전체'
distribution_20s['그룹'] = '20대'

distribution_combined = pd.concat([distribution_all, distribution_20s], axis=0)

distribution_combined['쇼핑업종'] = distribution_combined['쇼핑업종'].apply(lambda x: x.replace('_1순위쇼핑업종_', ''))

fig = px.bar(distribution_combined,
             x='쇼핑업종',
             y='비율',
             color='그룹',
             title='연체자의 1순위 쇼핑업종 비율 비교',
             labels={'비율': '비율 (%)', '쇼핑업종': '쇼핑업종'},
             barmode='group')

fig.update_layout(xaxis_tickangle=-45)
fig.show()
```
